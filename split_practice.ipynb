# **split 함수의 사용법**
spilt 함수는 말 그대로 나누는 함수, 그렇다면 무엇을 나누는가?

***문자열***을 ***리스트로*** ***나눠서*** 저장한다.

기본적인 사용법은 문자열을 저장한 변수에 `.split("나누고 싶은 기준")`을 붙여 사용한다.


```
alpabet = "a,b,c"
alpabet.split(",") # 쉼표를 기준으로 문자열을 나눠서 리스트에 저장

>>> ["a", "b", "c"]
```
문자열 바로 뒤에 적용하여 사용할 수 있다.


```
"G E T O U T".split(" ") #공백을 기준으로 나눠서 리스트에 저장

>>> ["G", "E", "T", "O", "U", "T"]
```

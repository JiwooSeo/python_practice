# **Integration with Simpson's Rule**


---
Simpson's Rule
Simpson's rule is a method for numerical integration, the numerical approximation of definite integrals. Specifically, it is the following approximation for ${\displaystyle n+1}$ values ${\displaystyle x_{0}\cdots x_{n}}$ ${\displaystyle x_{0}\cdots x_{n}}$ bounding ${\displaystyle n}$ equally spaced subdivisions **(where ${\displaystyle n}$ is even)**: (General Form)

$\int_a^bf(x)dx \approx S_n= ∆x[f(x_0 )+4f(x_1 )+2f(x_2 )+4f(x_3 )+⋯+2f(x_(n-2) )+4f(x_(n-1) )+f(x_n)]/3$




```
import numpy as np
import matplotlib.pyplot as plt
```

```
#divide interval [first, last] into n subintervals of equal length
step = 1e-4 
x = np.arange(0,np.pi,step)

n = len(x)
a = x[0]
b = x[-1]
dx = (b-a)/n #Δx

#sample point of x: x_i = first + i*Δx for all i
x_sam = []  
for i in range(0,n):
  if i == 0:
    x_sam.append(a)
  else:
    if x_sam[-1] <= b:
      xnext = a+i*dx
      x_sam.append(xnext)

#as above method, last number can be different with last number of interval
#append last number of interval
if x_sam[-1] != b: 
  x_sam.append(b)

#define fuction want to integrate
f_x = [] #set of value of fuction f(x)
for i in x_sam:
    fuc = np.sin(i) #define f(x)=e^x
    f_x.append(fuc) 
  
  
```

```
#This method is using idea of inducing Simpson's Rule

#set of area
A_i = []

if n%2 == 0: #because n must be even number
  for i in range(0,len(x_sam)):
    if i < len(x_sam)-2: #avoid the error: out of range of list
      a_i = dx*(f_x[i] + f_x[i+1]+ f_x[i+2])/3
      A_i.append(a_i)
    else:
      pass
else:
  for i in range(0,len(x_sam)+1):
      for i in range(0,len(x_sam)):
        if i < (len(x_sam)-2):
          a_i = dx*(f_x[i] + f_x[i+1]+ f_x[i+2])/3
          A_i.append(a_i)
        else:
          pass


def Integration_simpsons(x):
  sum_value = sum(A_i)
  return sum_value

```

```
print("Integration value with Simpson's Rule : ", Integration_simpsons(f_x))

plt.plot(x_sam, f_x)
plt.fill_between(x_sam, f_x, facecolor='skyblue', alpha=.8,)
plt.show()
```

#**Integration with Midpoint Rule**


---
Approximating f at the midpoint of subintervals $\ [x_i$ $ , x_(i+1)]: x_i^*= (x_i+x_(i+1))/2$, and calculate Rimman Sum
$\int_a^b f(x)dx \approx M_n =Δx[f(x_1^*)+f(x_2^*)+f(x_3^*)+...+f(x_n^*)] $

where $Δx=(b-a)/n  $


```
import numpy as np
import matplotlib.pyplot as plt
```

```
#divide interval [first, last] into n subintervals of equal length
step = 1e-4
x = list(np.arange(0,np.pi,step))
n = len(x)
a = x[0]
b = x[-1]

dx = (b-a)/n

#find midpoint of subinterval [x_i, x_(i+1)]
x_avg = []
for i in range(n):
  x_i = (x[i]+x[i-1])/2
  x_avg.append(x_i)

#define fuction f(x)
f_x = []
for i in x_avg:
  f_i= np.sin(i) # f(x) = e^x
  f_x.append(f_i) 

def Integration_midpoint(x):
  sum_value = dx*sum(x)
  return sum_value

#remove irregular number
del x_avg[0]
del f_x[0]
```

```
print("Integration value with Midpoint Rule : ", Integration_midpoint(f_x))

plt.plot(x_avg, f_x)

plt.fill_between(x_avg, f_x,  facecolor='skyblue', alpha=.8,)

plt.show()
```

# **Integration with Trapezoidal Rule**

---
The trapezoidal rule works by approximating the region under the graph of the function ${\displaystyle f(x)}$ as a trapezoid and calculating its area.

$\int_a^b f(x)dx \approx T_n =Δx[f(x_1)+f(x_2)+f(x_3)+...+f(x_n)]/2 $

where $Δx= (b-a)/n$  and $x_i=a+iΔx $


```
import numpy as np
import matplotlib.pyplot as plt
```

```
#divide interval [first, last] into n subintervals of equal length
step = 1e-4
x = np.arange(0,np.pi,step)
n = len(x)
a = x[0]
b = x[-1]

dx = (b-a)/n

#set sample point
#x_i = a+i*dx
x_sam = []
for i in range(len(x)):
  if i == 0:
    x_sam.append(x[0])
  else:
    if x_sam[-1] <= b:
      xnext = x[0]+i*dx
      x_sam.append(xnext)

#as above method, last number can be different with last number of interval
#append last number of interval     
if x_sam[-1] != b:
  x_sam.append(b)

#Define fuction f(x)
f_x = [] 
for i in x_sam:
  f_i =np.sin(i)+np.sin(2*i)+np.sin(4*i)+np.sin(6*i)+np.cos(i)+np.cos(3*i)+np.cos(5*i)+np.cos(7*i)

  f_x.append(f_i)

```

```
def Integration_Trapezoidal(x):
  partial_sum_set = []

  for i in range(1,len(x)):
    sum_i = x[i]+x[i-1]
    partial_sum_set.append(sum_i)
  sum_value = sum(partial_sum_set)
  intergration_value = sum_value*dx/2
  return intergration_value
```

```
print("Integration value with Trapezoidal Rule : ", Integration_Trapezoidal(f_x))
plt.plot(x_sam, f_x)
plt.fill_between(x_sam, f_x,  facecolor='skyblue', alpha=.8,)
plt.show()
```

# Gradient descent

```
def numerical_derivative_point(point,f):
  h = 1e-5
  derivative = (f(point+h)-f(point))/h
  return derivative

def gradiant_descent(x,f,step):
  a = f(x)
  ran = np.random.choice(x)
  lr = 0.5
  n = 0
  while n < step :
    if numerical_derivative_point(ran,f)>0:
      ran = ran - lr
    elif numerical_derivative_point(ran,f)<0:
      ran = ran +lr
    elif numerical_derivative_point(ran,f) ==0:
      break
    plt.plot(ran,f(ran),"o", color ="red")
    n = n+1
  plt.plot(x,a)
  plt.show()
  return ran

```

```
x = np.arange(-10,10,0.001)
def f(x):
  return x**3
print(gradiant_descent(x,f,1000))

```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return  np.sin(x)+np.sin(2*x)+np.sin(4*x)+np.sin(6*x)+np.cos(x)+np.cos(3*x)+np.cos(5*x)+np.cos(7*x)

start =-5*np.pi
stop = 5*np.pi

x = np.arange(start,stop,0.0001)
b = min(f(x))
h = max(f(x))
markersize = 1
count = []
y = f(x)
yy = np.arange(b,h,0.0001)

plt.figure(figsize=(10,5))
for i in range(100000):
  x_p = np.random.choice(x)
  y_p = np.random.choice(yy)


  if start<=x_p<=stop and (0<=y_p<=f(x_p)):
    count.append(1)
    plt.plot(x_p,y_p, "o", color = "red",ms=markersize)
  elif  start<=x_p<=stop and f(x_p)<=y_p<=0:
    count.append(-1)
    plt.plot(x_p,y_p, "o", color = "blue",ms=markersize)
  else:
    count.append(0)
    plt.plot(x_p,y_p, "o", color = "green",ms=markersize)

N_tot = len(count)
N = sum(count)

p =N/N_tot
a = (x[-1]-x[0])*(yy[-1]-yy[0])
A = p*a

print("percentage of point: ",p , "\ntotal area of domain: ",a,"\nIntegration value: ",A )

plt.plot(x,y)
plt.grid()
plt.show()
```
